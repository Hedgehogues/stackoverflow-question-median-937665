```python
import pandas as pd

data = '''1   5
1   4
1   3
1   2
1   4
2   5
2   6
2   8
2   3'''
data = [list(map(int, filter(lambda el: len(el) > 0, l.split(' ')))) for l in data.split('\n')]
data
```

```python
df = pd.DataFrame(data, columns=['com', 'x1'])
df
```

```python
df['delta'] = df['x1'].rolling(window=2).apply(lambda x: x[1] - x[0], raw=True)
df
```

```python
df['median'] = df['com']
df['median'] = df['median'].map(df.groupby('com')['delta'].median().to_dict())
df
```
